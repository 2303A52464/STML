```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
data = pd.read_csv('/content/Salary_dataset.csv')
data.head()
```

```python
x=data[['YearsExperience']]
y=data['Salary']
print(x)
print(y)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points') # Changed X to x
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y) # Changed X to x
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points') # Changed X to x
plt.plot(x, lin_reg.predict(x), color='red', label='Best Fit Line') # Changed X to x
plt.title('Years of Experience vs Salary with Best Fit Line')
plt.xlabel('Years of Experience')
```

```python
def evaluate_model(test_size):
  X_train, X_test, y_train, y_test = train_test_split(x, y, # Changed X to x
  test_size=test_size/100, random_state=42)
  lin_reg.fit(X_train, y_train)
  y_pred = lin_reg.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  return mse, mae
mse_20, mae_20 = evaluate_model(20)
mse_25, mae_25 = evaluate_model(25)
print(f'MSE for 20% test size: {mse_20:.2f}, MAE for 20% test size: {mae_20:.2f}')
print(f'MSE for 25% test size: {mse_25:.2f}, MAE for 25% test size: {mae_25:.2f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
data_dict = {
'Feature1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
'Feature2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
'Feature3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
'Feature4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
'Target': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]}
data = pd.DataFrame(data_dict)
data.head()
```

```python
x=data[['Feature1','Feature2','Feature3','Feature4']]
y=data['Target']
print(x)
print(y)
```

```python
plt.figure(figsize=(10,8))
for i, feature in enumerate(x.columns): # Changed X to x
    plt.subplot(2, 2, i+1)
    plt.scatter(x[feature], y, color='blue') # Changed X to x
    plt.title(f'{feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
plt.figure(figsize=(8,8))
for feature in x.columns:
    plt.scatter(x[feature], y, color='blue', label='Data points')
    plt.plot(x[feature], lin_reg.predict(x), color='red', label='Best Fit Line')
    plt.title(f'{feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Target')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def evaluate_model(test_size):
      X_train, X_test, y_train, y_test = train_test_split(x, y, # Fixed indentation and changed X to x
                                                          test_size=test_size/100, random_state=42)
      lin_reg = LinearRegression()
      lin_reg.fit(X_train, y_train)
      y_pred = lin_reg.predict(X_test)
      mse = mean_squared_error(y_test, y_pred)
      mae = mean_absolute_error(y_test, y_pred)
      return mse, mae
mse_20, mae_20 = evaluate_model(20)
mse_30, mae_30 = evaluate_model(30)
print(f'MSE for 20% test size: {mse_20:.2f}, MAE for 20% test size: {mae_20:.2f}')
print(f'MSE for 30% test size: {mse_30:.2f}, MAE for 30% test size: {mae_30:.2f}')
```
